```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):

  row = floor.shape[0]
  col = floor.shape[1]
  for i in range(row):   # 5 row, 6 col
      if i % 2 == 0:
          for j in range(0, col, 2):
              print(floor[i][j], end=" ")
      else:
          for j in range(col-1, -1, -2):
              print(floor[i][j], end=" ")
      print()




floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],
                  ['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],
                  ['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                  ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']]
                )
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
  row = m.shape[0]
  col = m.shape[1]

  result_arr = np.zeros((col, row), dtype="i")
  for i in range(col):
      for j in range(row):
          result_arr[i][j] = m[j][i]
      print()
  return result_arr


m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

```python
import numpy as np
def arrangement(m, diff):

    row = m.shape[0]
    col = m.shape[1]

    sub_arr = np.zeros((row-1), dtype="i")
 # shobgulo difference diff er shathe milte hobe. erjonno shob difference gulo store korte hobe jeno check kora jay shobgulor shathe.

    index = 0
    j = 0
    for i in range(row-1):
        j += 1
        sub_arr[index] = abs(m[i][j] - m[j][i])
        index += 1

    flag = True
    for i in sub_arr:
        if i != diff:
            flag = False
            break

    return flag


m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)

m = np.array([[8,15,7,12],[13,2,18,11],[9,20,5,2],[14,9,0,10]])
print_matrix(m)
print()
returned_value = arrangement(m,2)
print(returned_value)#This should print True
unittest.output_test(returned_value, True)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):

    i, j = knight

    if 0 > i or i > 7 or 0 > j or j > 7:
        return np.zeros((0, 0), dtype="i")

    result = np.zeros((8, 8), dtype="i")
    result[i][j] = 33

    pos_r1 = i + 2
    pos_c1 = j + 1
    if 0 <= pos_r1 <= 7 and 0 <= pos_c1 <= 7:
        result[pos_r1][pos_c1] = 3

    pos_r2 = i + 1
    pos_c2 = j + 2
    if 0 <= pos_r2 <= 7 and 0 <= pos_c2 <= 7:
        result[pos_r2][pos_c2] = 3

    pos_r3 = i + 2
    pos_c3 = j - 1
    if 0 <= pos_r3 <= 7 and 0 <= pos_c3 <= 7:
        result[pos_r3][pos_c3] = 3

    pos_r4 = i - 2
    pos_c4 = j + 1
    if 0 <= pos_r4 <= 7 and 0 <= pos_c4 <= 7:
        result[pos_r4][pos_c4] = 3

    pos_r5 = i + 1
    pos_c5 = j - 2
    if 0 <= pos_r5 <= 7 and 0 <= pos_c5 <= 7:
        result[pos_r5][pos_c5] = 3

    pos_r6 = i - 1
    pos_c6 = j + 2
    if 0 <= pos_r6 <= 7 and 0 <= pos_c6 <= 7:
        result[pos_r6][pos_c6] = 3

    pos_r7 = i - 2
    pos_c7 = j - 1
    if 0 <= pos_r7 <= 7 and 0 <= pos_c7 <= 7:
        result[pos_r7][pos_c7] = 3

    pos_r8 = i - 1
    pos_c8 = j - 2
    if 0 <= pos_r8 <= 7 and 0 <= pos_c8 <= 7:
        result[pos_r8][pos_c8] = 3

    return result


knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```
