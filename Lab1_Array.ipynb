```python
# Complete the functions defined in this cell


# Test 01: Play Right

sequence=[10,20,30,40,50,60]
beats = [1,0,0,1,0,1]

def playRight(sequence,beats):
# Right rotate
  for i in beats:
    if i == 1:
      temp = sequence[len(sequence) - 1]
      for i in range(len(sequence)-1, 0 , -1):
        sequence[i] = sequence[i-1]
      sequence[0] = temp

  return (sequence)

print("///  Test 01: Play Right  ///")
sequence=np.array([10,20,30,40,50,60])
beats = np.array([1,0,0,1,0,1])
returned_value = playRight(sequence, beats)
print(f'Task 1: {returned_value}') # This should print [40, 50, 60, 10, 20, 30]
unittest.output_test(returned_value, np.array([40, 50, 60, 10, 20, 30]))




# Test 02: Discard Cards  (last e or ache)

def removeElement(arr, index):
    for i in range(index+1, len(arr)):  # Left shift
        arr[i-1] = arr[i]
    arr[len(arr)-1] = 0   # None er jagay 0.


def discardCards(cards,number):
  i = 0
  while i <= len(cards)-1:
    if cards[i] == number:
      removeElement(cards, i)  # driver code call korar dorkar hoy na.
    else:
      i += 1

  # for loop diye korle ekta 2 theke jay keno jani!
  # for i in range(len(cards)):
  #   if cards[i] == number:
  #     removeElement(cards, i)

  return np.array(cards)


print("///  Test 02: Discard Cards  ///")
cards = np.array([1,2,3,2,8,2,2,5,7])
returned_value = discardCards(cards, 2)
print(f'Task 2: {returned_value}') # This should print [1, 3, 8, 5, 7, None, None, None, None]
unittest.output_test(returned_value, [1, 3, 8, 5, 7, 0, 0, 0, 0])




# Test 03: Merge Lineup

def mergeLineup(pokemon_1, pokemon_2):
  merged_arr = np.zeros((len(pokemon_1)), dtype = 'i')

  for i in range(len(pokemon_1)):
    if pokemon_1[i] == None:
      pokemon_1[i] = 0

  for i in range(len(pokemon_2)):
    if pokemon_2[i] == None:
      pokemon_2[i] = 0


  i = 0
  j = len(pokemon_1)-1

  while i <= len(pokemon_1)-1:
  # for k in range(len(pokemon_1)): for loop eo hoy. tokhon merged_arr[k] hobe.
    merged_arr[i] = pokemon_1[i] + pokemon_2[j]
    i += 1
    j -= 1

  return merged_arr

print("///  Test 03: Merge Lineup  ///")
pokemon_1 = np.array([12, 3, 25, 1, None])
pokemon_2 = np.array([5, -9, 3, None, None] )
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}') # This should print [12, 3, 28, -8, 5]
unittest.output_test(returned_value, np.array([12, 3, 28, -8, 5]))

pokemon_1 = np.array([4, 5, -1, None, None])
pokemon_2 = np.array([2, 27, 7, 12, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}') # This should print [4,17,6,27,2]
unittest.output_test(returned_value, np.array([4,17,6,27,2]))

pokemon_1 = np.array([4, 5, None, None])
pokemon_2 = np.array([2, None, None, None])
returned_value =mergeLineup(pokemon_1, pokemon_2)
print(f'Task 3: {returned_value}') # This should print [4,5,0,2]
unittest.output_test(returned_value, np.array([4,5,0,2]))



# Test 04: Balance your Salami

def balanceSalami(salami):
    total_salami = 0
    for i in salami:
        total_salami += i

    if total_salami % 2 != 0:
        return False
    else:
        my_salami = total_salami // 2
        sib_salami = 0

        for i in salami:
            sib_salami += i
            if sib_salami == my_salami:    # 1,1,1,2,1 = 6
                return True
            elif sib_salami > my_salami:   # 1,1,2,1,1 = 6  [notes gulor serial change kora jabe na]
                return False

print("///  Test 04: Balance Your Salami  ///")
salami = np.array([1, 1, 1, 2, 1])
returned_value = balanceSalami(salami)
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)

salami = [2, 1, 1, 2, 1]
returned_value = np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print False
unittest.output_test(returned_value, False)

salami = [10, 3, 1, 2, 10]
returned_value =  np.array(balanceSalami(salami))
print(f'Task 4: {returned_value}') # This should print True
unittest.output_test(returned_value, True)





# Test 05: Protecc Salami

def protectSalami(salami):

  max = 0
  for i in range(len(salami)):
      if salami[i] >= max:
          max = salami[i]

  new_arr = np.zeros ((max + 1), dtype = 'i')  #max+1 karon index 0 theke start.
  for i in salami:
      new_arr[i] = new_arr[i] + 1

  flag = False
  for i in range(len(new_arr)):
      if new_arr[i] > 1:
          for j in range(len(new_arr)):
              if new_arr[i] == new_arr[j]:
                  flag = True
              else:
                  flag = False
  return flag

print("///  Test 05: Protecc Salami  ///")
salami = np.array([4,5,6,6,4,3,6,4])
returned_value = protectSalami(salami)
print(f'Task 5: {returned_value}') # This should print True
unittest.output_test(returned_value, True)

salami = np.array([3,4,6,3,4,7,4,6,8,6,6])
returned_value = protectSalami(salami)
print(f'Task 5: {returned_value}') # This should print False
unittest.output_test(returned_value, False)






# Test 06: Odd Even Wave

def waveYourFlag(arr):
    count_even = 0
    count_odd = 0
    for i in arr:
        if i % 2 == 0:
            count_even += 1
        else:
            count_odd += 1

    even = [None] * count_even
    odd = [None] * count_odd
    result = [None] * len(arr)
    even_index = 0
    odd_index = 0
    result_index = 0

    for i in arr:
        if i % 2 == 0:
            even[even_index] = i
            even_index += 1
        else:
            odd[odd_index] = i
            odd_index += 1

    i = 0
    j = 0
    if arr[i] % 2 == 0:  #first elem even hole
        for i in range(len(even)):
            result[result_index] = even[i]
            result_index += 1

            if len(odd) > i:  #[2,12,8] & [3,1,5] ekhane ei condition na dileo hoto karon duitar len same
                result[result_index] = odd[j]
                result_index += 1
                j += 1

    else:  # first elem odd hole
        for i in range(len(odd)):
            result[result_index] = odd[i]
            result_index += 1

            if len(even) > i:  #[45,23,41] & [78,84] --must-- Ei condition na dile 84 er por arekta elem ke khujto.
                result[result_index] = even[j]       #first e je elem diye start hoy(even/odd)oitar len ta arektar equal thake or boro thake. so, ei duirokom case e possible.
                result_index += 1
                j += 1

    return np.array(result)


print("///  Test 06: Odd Even Wave  ///")
arr = np.array([2,12,3,8,1,5])
returned_value = waveYourFlag(arr)
print(f'Task 6: {returned_value}') # This should print [2,3,12,1,8,5]
unittest.output_test(returned_value, np.array([2,3,12,1,8,5]))

arr = np.array([45,23,78,84,41])
returned_value = waveYourFlag(arr)
print(f'Task 6: {returned_value}') # This should print [45,78,23,84,41]
unittest.output_test(returned_value, np.array([45,78,23,84,41]))
```

```python
# Part 2

def mean_value(array):
    sum = 0
    for i in array:
        sum += i
    mean = sum/len(array)
    return mean



def standard_deviation(array):
    summation = 0
    for i in array:
        summation += (i - mean_value(array))**2
    standardDeviation = round( ( summation / (len(array)-1) ) ** 0.5 , 2)
    return standardDeviation



def new_array(array):
    newArray = [None] * len(array)
    away1_below = mean_value(array) - (1.5 * standard_deviation(array))
    away2_above = mean_value(array) + (1.5 * standard_deviation(array))
    # print(away1_below)
    # print(away2_above)

    index = 0
    count = 0
    for i in array:
        if i < away1_below or i > away2_above:
            newArray[index] = i
            index += 1
            count += 1   #count = resultArray er size(None baade koyta elem ase)


    resultArray = [None] * count
    for i in range(len(newArray)):
        if newArray[i] != None:
            resultArray[i] = newArray[i]

    if count > 0:
        return resultArray
    else:
        return None

array = [10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]
print("The mean of the numbers is:", mean_value(array))
print("The standard deviation is:", standard_deviation(array))
print("New array:", new_array(array))
```
