```python
class Patient:

    def __init__(self, id, name, age, bloodgroup, next, prev):
        self.id = id
        self.name = name
        self.age = age
        self.bloodgroup = bloodgroup
        self.next = next
        self.prev = prev

class WRM:
    def __init__(self):
        self.head = Patient(None, None, None, None, None, None)
        self.head.next = self.head
        self.head.prev = self.head

    def RegisterPatient(self, id, name, age, bloodgroup):
        new_patient = Patient(id, name, age, bloodgroup, None, None)
        last = self.head.prev
        last.next = new_patient
        new_patient.prev = last
        new_patient.next = self.head
        self.head.prev = new_patient

    def ServePatient(self):
        self.head.next = self.head.next.next
        self.head.next.next.prev = self.head

    def CancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head

    def CanDoctorGoHome(self):
        self.flag = True
        if self.head.next == self.head:
          return self.flag
        else:
          self.flag = False
          return self.flag

    def ShowAllPatient(self):
        temp = self.head.next
        while temp != self.head:
            print(temp.id, end =" ")
            temp = temp.next
        print()


l = WRM()
l.ServePatient()
l.CancelAll()
l.CanDoctorGoHome()
l.ShowAllPatient()

while True:
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        l.RegisterPatient(id, name, age, blood_group)
        print("Patient added successfully!")

    elif choice == 2:
        print("Patient being served:", l.head.next.name)  # (self.variable) ke driver code e (obj.variable) likhte hoy.

    elif choice == 3:
        print("All patients in sequence to serve:")
        l.ShowAllPatient()

    elif choice == 4:
        l.CanDoctorGoHome()
        if l.flag == True:
            print("Yes")
        else:
            print("No")

    elif choice == 5:
        l.CancelAll()
        print("All appointments canceled successfully!")

    elif choice == 6:
       break

    print()
```
